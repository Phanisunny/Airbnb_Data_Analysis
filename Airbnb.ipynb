```python
import pandas as pd
air=pd.read_csv("Airbnb_Open_Data.csv")
air.isnull().sum()
```

```python
def bubble(a,i=0,j=1):
    if(i==len(a)-1):
        return a
    elif(j==len(a)):
        return bubble(a,i+1,i+2)
    else:
        if(a[i]>a[j]):
            temp=a[i]
            a[i]=a[j]
            a[j]=temp
        return bubble(a,i,j+1)
```
